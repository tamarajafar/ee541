```python
# n is how many uniform(0,1) draws does it take until the sum exceeds 4
import random

def sample_N(threshold=4.0):
    total = 0.0
    n = 0
    while total <= threshold:
        total += random.random()
        n += 1
    return n
```

```python
def simulate_many_N(num_trials, threshold=4.0):
    return [sample_N(threshold) for _ in range(num_trials)]
```

```python
import matplotlib.pyplot as plt

for trials in [100, 1000, 10000]:
    Ns = simulate_many_N(trials, threshold=4.0)

    plt.figure()
    plt.hist(Ns, bins=range(min(Ns), max(Ns)+2), edgecolor='black', align='left')
    plt.title(f'Histogram of N (threshold=4), {trials} trials')
    plt.xlabel('N = # Uniform(0,1) draws needed for sum > 4')
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
# as the trials increase the hsitograms become smoother and more stable
```

```python
for trials in [100, 1000, 10000]:
    Ns = simulate_many_N(trials, threshold=4.0)
    print(trials, "mean(N) =", sum(Ns)/len(Ns))
```

```python
# the n stabilizes as the number of trials increase
```
