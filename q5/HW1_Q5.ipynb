```python
import random

def flip(p_head=0.7):
    return 1 if random.random() < p_head else 0
    # 1 = HEAD, 0 is tails
```

```python
def simulate_flips(n, p=0.7):
    flips = []
    for _ in range(n):
        flips.append(flip(p))
    return flips
```

```python
flips_50 = simulate_flips(50, 0.7)
```

```python
flips_50
```

```python
heads = sum(flips_50)
heads
```

```python
def longest_run_of_heads(flips):
    current = 0
    longest = 0
    for outcome in flips:
        if outcome == 1:     # heads
            current += 1
            if current > longest:
                longest = current
        else:                # tail breaks the run
            current = 0
    return longest
```

```python
longest_50 = longest_run_of_heads(flips_50)
longest_50
```

```python
def repeated_head_counts(num_experiments, n_flips=50, p=0.7):
    counts = []
    for _ in range(num_experiments):
        flips = simulate_flips(n_flips, p)
        counts.append(sum(flips))
    return counts
```

```python
import matplotlib.pyplot as plt

for N in [20, 100, 200, 1000]:
    counts = repeated_head_counts(N, 50, 0.7)

    plt.figure()
    plt.hist(counts, bins=range(0, 52), edgecolor='black')
    plt.title(f'Heads in 50 flips, repeated {N} times (p=0.7)')
    plt.xlabel('Number of heads out of 50')
    plt.ylabel('Frequency')
    plt.show()
```

```python
#as we did more flips the histogram became a miuch smoother distribution
```

```python
def head_run_lengths(flips):
    runs = []
    current = 0

    for outcome in flips:
        if outcome == 1:
            current += 1
        else:
            if current > 0:
                runs.append(current)
                current = 0

    if current > 0:
        runs.append(current)

    return runs
```

```python
flips_500 = simulate_flips(500, 0.7)
runs = head_run_lengths(flips_500)

plt.figure()
plt.hist(runs, bins=range(1, max(runs)+2), edgecolor='black')
plt.title('Histogram of heads run lengths (500 flips, p=0.7)')
plt.xlabel('Run length (consecutive heads)')
plt.ylabel('Frequency')
plt.show()
```
